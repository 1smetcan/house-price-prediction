# House Price Prediction Using Regression Models

### Introduction
In this project, I aim to predict house prices using various machine learning models. I will explore the dataset to understand the key features and relationships between variables, perform feature engineering, and build regression models to predict house prices. 

### Import Necessary Libraries

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
```

#### Setting Display Options

```python
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", 100)
```

## Exploratory Data Analysis and Feature Engineering for Training Data

```python
df = pd.read_csv("train.csv")
```

#### Separating Numeric and Categoric Columns

```python
categoric_columns = [i for i in df.columns if df[i].dtype == "object"]
df_numeric = df.drop(columns=categoric_columns, axis=1)
df_categoric = df[categoric_columns]
```

### Data Preprocessing for Numeric Features

```python
df_numeric.head()
```

#### Identifying Important Numeric Features Based on Correlation with SalePrice

```python
correlation = df.corr(numeric_only=True)["SalePrice"]
important_numeric_columns = correlation[abs(correlation) > 0.20].index.tolist()
print(important_numeric_columns)
```

```python
df_important_numeric = df_numeric[important_numeric_columns]
```

```python
df_important_numeric.head()
```

```python
df_important_numeric.isnull().sum()
```

```python
df_important_numeric["LotFrontage"]
```

```python
df_important_numeric["GarageYrBlt"]
```

#### Handling Missing Data

```python
imputer = SimpleImputer(strategy="most_frequent")
df_important_numeric = pd.DataFrame(imputer.fit_transform(df_important_numeric), columns=df_important_numeric.columns)
```

```python
df_important_numeric.info()
```

### Data Preprocessing for Categorical Features

```python
df_categoric.head()
```

```python
df_categoric.nunique()
```

```python
for i in df_categoric.columns:
    print(i, df[i].unique())
```

```python
df_categoric.isnull().sum()
```

#### One-Hot Encoding of Categorical Features

```python
df_categoric = pd.get_dummies(df_categoric, drop_first=True)
```

```python
df_categoric.head()
```

```python
df_categoric.shape
```

```python
df_categoric["SalePrice"] = df["SalePrice"]
```

#### Identifying Important Categoric Features Based on Correlation with SalePrice

```python
correlation2 = df_categoric.corr(numeric_only=True)["SalePrice"]
important_categoric_columns = correlation2[abs(correlation2) > 0.20].index.tolist()
print(important_categoric_columns)
```

```python
df_important_categoric = df_categoric[important_categoric_columns]
```

```python
df_important_categoric.head()
```

#### Merging Categorical and Numeric Features to Create Final DataFrame

```python
df_important_numeric = df_important_numeric.drop("SalePrice", axis=1)
last_df = df_important_numeric.join(df_important_categoric)
last_df.head()
```

## Model Training and Evaluation


#### Splitting Training Data into Features (X) and Target (Y)

```python
x = last_df.drop("SalePrice", axis=1)
y = last_df[["SalePrice"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=.20, random_state=42)
```

#### Building a Linear Regression Model with Training Data

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

#### Testing Linear Regression Model

```python
mse = mean_squared_error(y_test, y_pred) ** .5
print(mse)
r2 = r2_score(y_test, y_pred)
print(r2)
```

#### Building a Random Forest Regression Model with Training Data

```python
rfr = RandomForestRegressor()
rfr.fit(x_train, y_train)
y_pred2 = rfr.predict(x_test)
```

#### Testing Random Forest Regression Model

```python
mse2 = mean_squared_error(y_test, y_pred2) ** .5
print(mse2)
r2_2 = r2_score(y_test, y_pred2)
print(r2_2)
```

#### Enhancing the Performance of Random Forest Regression Model with XGBoost

```python
rfr_model = xgb.XGBRFRegressor(
    n_estimators=200,
    max_depth=10, 
    learning_rate=1,
    subsample=0.8, 
    colsample_bynode=0.8,
    random_state=42
)
rfr_model.fit(x_train, y_train)
y_pred3 = rfr_model.predict(x_test)
```

```python
mse3 = mean_squared_error(y_test, y_pred3) ** .5
print(mse3)
r2_3 = r2_score(y_test, y_pred3)
print(r2_3)
```
